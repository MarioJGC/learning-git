# ¿Qué podemos entender con `.describe()`?

El método `.describe()` nos permite obtener un resumen estadístico rápido de los datos y sacar conclusiones básicas.

## 1. Cantidad y calidad de datos
- `count`: indica cuántos valores hay.
- Si `count` es menor al total de filas, hay valores faltantes (NaN).
- Permite comparar qué columnas tienen peor calidad de datos.

## 2. Tendencia central
- `mean` (promedio) y `50%` (mediana).
- Si mean ≈ mediana → datos simétricos.
- Si mean > mediana → sesgo a la derecha.
- Si mean < mediana → sesgo a la izquierda.

## 3. Dispersión
- `std`: mide qué tan dispersos están los datos.
- `std` alto → valores muy variables.
- `std` bajo → valores más estables.

## 4. Valores atípicos (outliers)
- Comparando `min`, `max`, `25%` y `75%` se pueden detectar posibles outliers.
- Valores muy alejados de los cuartiles suelen indicar errores o casos extremos.

## 5. Distribución de los datos
- Cuartiles (`25%`, `50%`, `75%`) muestran cómo se concentran los datos.
- Cuartiles cercanos → datos concentrados.
- Cuartiles separados → datos dispersos.

## 6. Variables categóricas
- `unique`: número de valores distintos.
- `top`: valor más frecuente.
- `freq`: frecuencia del valor más común.
- Ayuda a detectar sesgos o problemas de normalización.

## 7. Comparaciones posibles
- Comparar columnas entre sí.
- Analizar grupos con `groupby().describe()`.
- Evaluar calidad, estabilidad y distribución de datos.

## 8. Limitaciones
- No indica causalidad.
- No predice resultados.
- No muestra relaciones entre variables.

## Resumen
Con `.describe()` podemos:
- Detectar problemas en los datos.
- Entender el comportamiento de las variables.
- Identificar outliers.
- Comparar consistencia y dispersión.
